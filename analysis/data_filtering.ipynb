```python
import bz2
import json
```

```python
import pandas as pd
import numpy as np
```

```python
df_house_115 = pd.read_csv(r'/content/drive/MyDrive/ADA/house_115.csv')
df_senate_115 = pd.read_csv(r'/content/drive/MyDrive/ADA/senate_115.csv')
df_house_116 = pd.read_csv(r'/content/drive/MyDrive/ADA/house_116.csv')
df_senate_116 = pd.read_csv(r'/content/drive/MyDrive/ADA/senate_116.csv')
```

```python
df_house = pd.merge(df_house_115, df_house_116, how='inner', on=['fec_candidate_id'])
df_senate = pd.merge(df_senate_115, df_senate_116, how='inner', on=['fec_candidate_id'])
print(df_house)
print(df_senate)
```

```python
df_congress = pd.concat([df_house, df_senate], ignore_index=True)
```

```python
df_congress.groupby("party_x").party_x.agg(["count"])
```

```python
df_congress["speaker"] = (df_congress["first_name_x"] + " " + df_congress["last_name_x"]).str.lower()
```

```python
df_congress["speaker"]
```

```python
df_congress.to_csv('/content/drive/MyDrive/ADA/congress_115_116.csv')  
```

```python
%%time

path_to_file = '/content/drive/MyDrive/Quotebank/quotes-2020.json.bz2' 
path_to_out = '/content/drive/MyDrive/ADA/quotes-2020-congress-speakers.json.bz2'


with bz2.open(path_to_file, 'rb') as s_file:
  with bz2.open(path_to_out, 'wb') as d_file:
    for i, quote in enumerate(s_file):
      quote = json.loads(quote) # loading a quote
      if str(quote["speaker"]).replace(' ','').isalpha() and str(quote["speaker"]) != 'None':
        print(str(quote["speaker"]).lower(), i)
        if df_congress.loc[df_congress["speaker"].str.contains(str(quote["speaker"]).lower(), case=False)].size > 0 :
          d_file.write((json.dumps(quote)+'\n').encode('utf-8')) # writing in the new file
```
